```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Ethical criteria (axes)
labels = [
    "Proceeds Clarity", 
    "Issuance Reporting", 
    "Verification", 
    "Contingency", 
    "Greenwashing Risk"
]
num_vars = len(labels)

# Scores for each issuer type (scale: 05)
sovereign = [4, 3, 5, 3, 2]
financial_corp = [2, 2, 2, 1, 4]
dev_bank = [5, 4, 5, 3, 1]
supranational = [4, 4, 5, 2, 2]

# Close the radar chart loop
def wrap(values):
    return values + values[:1]

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

# Initialize radar plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot each issuer
ax.plot(angles, wrap(sovereign), label='Sovereign', marker='o')
ax.plot(angles, wrap(financial_corp), label='Financial Corp.', marker='s')
ax.plot(angles, wrap(dev_bank), label='Development Bank', marker='^')
ax.plot(angles, wrap(supranational), label='Supranational', marker='d')

# Optional: fill area
ax.fill(angles, wrap(sovereign), alpha=0.1)
ax.fill(angles, wrap(financial_corp), alpha=0.1)
ax.fill(angles, wrap(dev_bank), alpha=0.1)
ax.fill(angles, wrap(supranational), alpha=0.1)

# Set axis labels and title
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=10)
ax.set_yticklabels([])  # Hide radial ticks
ax.set_title("Ethical Disclosure Radar (EDR): Issuer Profile Comparison", fontsize=11, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
xtick_labels = ax.get_xticklabels()
xtick_labels[0].set_position((1.5, -0.2))
xtick_labels[2].set_position((1.5, -0.1))
xtick_labels[3].set_position((1.5, -0.1))
xtick_labels[4].set_position((1.5, -0.1))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Timeline events and years
events = [
    "NAICM Green Bonds Issued\n($6B under ICMA)", 
    "Construction Progresses", 
    "Referendum to Cancel Airport", 
    "Project Cancellation Begins", 
    "Bond Downgrade Begins", 
    "Government Offers Discounted Buyback", 
    ]

years = [2016, 2017, 2018.1, 2018.8, 2018.9, 2019.0, 2019.1]

# Plot setup
fig, ax = plt.subplots(figsize=(12, 3))
ax.hlines(1, min(years)-0.1, max(years)+0.1, color='gray', linestyle='--')
ax.set_ylim(0.9, 1.1)
ax.set_yticks([])

# Event markers
for i, (year, label) in enumerate(zip(years, events)):
    color = 'red' if "No Legal" in label else 'blue'
    ax.plot(year, 1, 'o', color=color)
    ax.text(year, 1.02, label, rotation=45, 
            ha='right' if i % 2 == 0 else 'left', fontsize=9)

# Axis and title
ax.set_xticks(range(2016, 2020))
ax.set_title("Mexico City Airport Green Bond: Timeline of Default & Investor Fallout (20162019)", 
fontsize=11, pad=80)
ax.set_xlabel("Year")

plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go

# Define the nodes
labels = [
    "Issuer Promises",            # 0
    "Voluntary Framework", # 1
    "External Review", # 2
    "Listed on Exchange",         # 3
    "Proceeds Deployed",          # 4
    "Project Underperforms",      # 5
    
    "Investor Loss"               # 7
]

# Define flow: from source node to target node
source = [0, 1, 2, 3, 4, 5, 6]
target = [1, 2, 3, 4, 5, 6, 7]
values = [90, 80, 60, 40, 30, 20, 10]  # Decreasing values to show accountability erosion

# Create Sankey diagram
fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=10,
        line=dict(color="black", width=0.5),
        label=labels,
        color="lightgreen"
    ),
    link=dict(
        source=source,
        target=target,
        value=values,
        color="rgba(100, 100, 255, 0.4)"
    )
)])

# Add title and show
fig.update_layout(title_text="Green Bond Lifecycle: Disappearing Legal Accountability", font_size=19)
fig.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Labels and values for each stage
labels = [
    "Initial Value",
    "Post-Issuance Market",
    "Project Underperformance",
    "No Exit Mechanism",
    "Final Investor Value"
]

# Values representing the change at each stage
changes = [100, -10, -30, -20, -40]

# Calculate running totals
values = [changes[0]]
for change in changes[1:]:
    values.append(values[-1] + change)

# Set up waterfall bar colors
colors = ['green' if val >= 0 else 'red' for val in changes] + ['grey']

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(range(len(values)), values, color=colors[:-1])

# Annotate the bars with value labels
for i, value in enumerate(values):
    ax.text(i, value + 0.5, f'{value}%', ha='center', va='bottom', fontsize=10)


# Formatting
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.set_ylabel("Investor Value (%)")
ax.set_title("Waterfall Chart: Investor Value Erosion in Failed Green Bonds")

# Format y-axis to show percentage
ax.yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{int(y)}%'))

plt.tight_layout()
plt.show()
```
